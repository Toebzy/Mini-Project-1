### Explore and clean

```python
ytexcel.isnull().sum()
```

```python
ytexcel = ytexcel.drop('Content Type', axis=1)
```

```python
ytexcel["Category"] = ytexcel["Category"].fillna("Mixed")
ytexcel.head()
```

```python
ytexcel = ytexcel.drop(0)
ytexcel.reset_index()
```

```python
ytexcel["Country"] = ytexcel["Country"].fillna("International")
ytexcel.head()
```

```python
#making the M, K into FLOATS and not strings
ytexcel['Subscribers'] = pd.to_numeric(ytexcel['Subscribers'].str.replace('M', ''), errors='coerce') * 1000000
```

```python
def parse_subscribers(value):
    if 'M' in value:
        return float(value.replace('M', '')) * 1000000
    elif 'K' in value:
        return float(value.replace('K', '')) * 1000
    else:
        return float(value)
```

```python
ytexcel['Average Views'] = ytexcel['Average Views'].apply(parse_subscribers)
ytexcel['Average Likes'] = ytexcel['Average Likes'].apply(parse_subscribers)
ytexcel['Average Comments'] = ytexcel['Average Comments'].apply(parse_subscribers)
```

```python
carscsv = carscsv.drop(['Market Category'], axis=1)
```

```python
mean_hp=carscsv['Engine HP'].mean()
carscsv['Engine HP']= carscsv['Engine HP'].fillna(mean_hp)
```

```python
carscsv.head()
```

```python
carscsv_cylin = carscsv[pd.isna(carscsv['Engine Cylinders'])]
carscsv_cylin.head()
```

```python
carscsv['Engine Cylinders'] = carscsv['Engine Cylinders'].fillna(0)
carscsv.head()
```

```python
carscsv.isnull().sum()
```

```python
carscsv_doors = carscsv[pd.isna(carscsv['Number of Doors'])]
carscsv_doors.head(8)
```

```python
# looked up the data we needed
condition = (carscsv['Make'] == 'Tesla') & (carscsv['Model'] == 'Model S') & pd.isna(carscsv['Number of Doors'])
carscsv.loc[condition, 'Number of Doors'] = 5
condition = (carscsv['Make'] == 'Ferrari') & (carscsv['Model'] == 'FF') & pd.isna(carscsv['Number of Doors'])
carscsv.loc[condition, 'Number of Doors'] = 3
```

```python
carscsv.isnull().sum()
```

```python
carscsv_fueltype = carscsv[pd.isna(carscsv['Engine Fuel Type'])]
carscsv_fueltype.head()
```

```python
carscsv['Engine Fuel Type'] = carscsv['Engine Fuel Type'].fillna('regular unleaded')
```
